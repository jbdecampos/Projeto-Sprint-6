````python
#Preenchimento dos valores ausentes em 'year_of_release' com base nos dados de 'name'

name_to_year = df.dropna(subset=['year_of_release']).set_index('name')['year_of_release'].to_dict()
def fill_year(row):
    if pd.isna(row['year_of_release']):
        return name_to_year.get(row['name'], row['year_of_release'])
    return row['year_of_release']

df['year_of_release'] = df.apply(fill_year, axis=1)

# Verifique se os valores foram preenchidos corretamente
print(df['year_of_release'].isna().sum())
````

````python
#Calculo do Ciclo de vida médio das plataformas 

platform_lifecycle = df.groupby('platform')['year_of_release'].agg(['min', 'max'])
platform_lifecycle['lifecycle_duration'] = platform_lifecycle['max'] - platform_lifecycle['min']


print(platform_lifecycle)

# Calculo da média do ciclo de vida
average_lifecycle = platform_lifecycle['lifecycle_duration'].mean()
print(f"Ciclo de vida médio das plataformas: {average_lifecycle:.2f} anos")

# Calculo da mediana do ciclo de vida
average_lifecycle = platform_lifecycle['lifecycle_duration'].median()
print(f"Ciclo de vida mediano das plataformas: {average_lifecycle:.2f} anos")
````

````python
#Calculo do Ciclo de vida médio das plataformas mais relevantes

platform_lifecycle_top = df_top_platforms.groupby('platform')['year_of_release'].agg(['min', 'max'])
platform_lifecycle_top['lifecycle_duration'] = platform_lifecycle_top['max'] - platform_lifecycle_top['min']
print(platform_lifecycle_top)

# Calculando a média do ciclo de vida
average_lifecycle_top = platform_lifecycle_top['lifecycle_duration'].mean()
print(f"Ciclo de vida médio das plataformas mais relevantes: {average_lifecycle_top:.2f} anos")

# Calculando a média do ciclo de vida
average_lifecycle_top = platform_lifecycle_top['lifecycle_duration'].median()
print(f"Ciclo de vida mediano das plataformas mais relevantes: {average_lifecycle_top:.2f} anos")
````

````python
#Tempo até o pico de venda de todas as plataformas 
platform_peak = df.groupby(['platform', 'year_of_release'])['total_sales'].sum().groupby(level=0).idxmax()
platform_lifecycle['time_to_peak'] = platform_peak.apply(lambda x: x[1]) - platform_lifecycle['min']

print(platform_lifecycle[['lifecycle_duration', 'time_to_peak']])

# Calculando a média do tempo até o pico
average_time_to_peak = platform_lifecycle['time_to_peak'].mean()
print(f"Tempo médio até o pico de vendas: {average_time_to_peak:.2f} anos")

# Calculando a mediana do tempo até o pico
average_time_to_peak = platform_lifecycle['time_to_peak'].median()
print(f"Tempo mediano até o pico de vendas: {average_time_to_peak:.2f} anos")
````

````python
#Tempo até pico de vendas das plataformas mais relevantes

platform_peak_top = df_top_platforms.groupby(['platform', 'year_of_release'])['total_sales'].sum().groupby(level=0).idxmax()
platform_lifecycle_top['time_to_peak'] = platform_peak_top.apply(lambda x: x[1]) - platform_lifecycle_top['min']

print(platform_lifecycle_top[['lifecycle_duration', 'time_to_peak']])

# Calculando a média do tempo até o pico
average_time_to_peak_top = platform_lifecycle_top['time_to_peak'].mean()
print(f"Tempo médio até o pico de vendas das plataformas mais relevantes: {average_time_to_peak_top:.2f} anos")

# Calculando a mediana do tempo até o pico
average_time_to_peak_top = platform_lifecycle_top['time_to_peak'].mean()
print(f"Tempo mediano até o pico de vendas das plataformas mais relevantes: {average_time_to_peak_top:.2f} anos")

````

### Etapa 4 - Criação de perfil por região

````python
# Função para criar o perfil das regiões 

def create_region_profile(region_column):
    profile = {}
    
    # Total de vendas por região
    total_sales = df[region_column].sum()
    profile['Total Sales'] = total_sales
    
    # Top 5 plataformas por vendas na região
    top_platforms = df.groupby('platform')[region_column].sum().sort_values(ascending=False).head(5)
    profile['Top 5 Platforms'] = top_platforms.index.tolist()
    
    # Top 5 gêneros por vendas na região
    top_genres = df.groupby('genre')[region_column].sum().sort_values(ascending=False).head(5)
    profile['Top 5 Genres'] = top_genres.index.tolist()
    
    # Média das avaliações dos críticos
    if df['critic_score'].notnull().any():
        mean_critic_score = df['critic_score'].mean()
        profile['Mean Critic Score'] = mean_critic_score
    
    # Média das avaliações dos usuários
    df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')  # Converter para numérico, ignorando erros
    if df['user_score'].notnull().any():
        mean_user_score = df['user_score'].mean()
        profile['Mean User Score'] = mean_user_score
    
    return profile

# Criar perfis para cada região
na_profile = create_region_profile('na_sales')
eu_profile = create_region_profile('eu_sales')
jp_profile = create_region_profile('jp_sales')

print("NA Region Profile:")
print(na_profile)
print("\nEU Region Profile:")
print(eu_profile)
print("\nJP Region Profile:")
print(jp_profile)
````

````python
# Classificações de usuários para Xbox One e PC com o DF original não filtrado 

# tratando os dados
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
df = df.dropna(subset=['user_score'])

# Classificações de usuários para Xbox One e PC
xbox_scores = df[df['platform'] == 'XOne']['user_score']
pc_scores = df[df['platform'] == 'PC']['user_score']

# Teste t para Xbox One vs PC
t_stat, p_val = stats.ttest_ind(xbox_scores, pc_scores)
print(f"Teste t para Xbox One vs PC: t = {t_stat:.2f}, p-valor = {p_val:.5f}")

# Dados de exemplo: Classificações de usuários para gêneros Action e Sports
action_scores = df[df['genre'] == 'Action']['user_score']
sports_scores = df[df['genre'] == 'Sports']['user_score']

# Teste t para Action vs Sports
t_stat, p_val = stats.ttest_ind(action_scores, sports_scores)
print(f"Teste t para Action vs Sports: t = {t_stat:.2f}, p-valor = {p_val:.5f}")

# Interpretando o resultado
alpha = 0.05
if p_val < alpha:
    print("Rejeitamos a hipótese nula: há uma diferença estatisticamente significativa entre as médias.")
else:
    print("Não rejeitamos a hipótese nula: não há diferença estatisticamente significativa entre as médias.")
````

#### Conclusão dos testes de hipóteses 1:

Adotei α = 0,05 como nível de significância padrão.

Com isso, concluí que:

O valor negativo do teste t para Xbox One vs PC sugere que a média das classificações dos usuários para a plataforma Xbox One é menor que a média para a plataforma PC. O p-valor é extremamente baixo, levando à rejeição da hipótese nula de que as classificações médias dos usuários para as duas plataformas sejam iguais. Portanto, há uma diferença estatisticamente significativa entre as médias das classificações para Xbox One e PC.

Por outro lado, o teste t realizado para comparar as classificações dos gêneros Action e Sports usando o DataFrame completo resultou em um valor t positivo, mas o p-valor de 0.10407 foi maior que o nível de significância adotado. Isso significa que não rejeitamos a hipótese nula, indicando que não há evidências estatísticas suficientes para afirmar que as classificações médias dos usuários para os gêneros Action e Sports são diferentes.
